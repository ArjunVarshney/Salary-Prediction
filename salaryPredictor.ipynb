```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv")
```

```python
data.info()
```

```python
data.describe()
```

```python
data.drop("name",axis=1,inplace=True)
```

```python
all_skills= []
for i in range(len(data)):
   s = data.iloc[i]["skills"]
   s_arr = s.split("} , {")
   for skill in s_arr:
      if(skill not in all_skills):
         all_skills.append(skill)
```

```python
data.drop("skills",axis=1,inplace=True)
data
```

```python
x = data.drop("salary(in rs)",axis=1, inplace=False)
y = data["salary(in rs)"]
```

```python
# salary_column = 'salary(in rs)'

# for column in data.columns:
#     if column != salary_column:
#         plt.scatter(data[column], data[salary_column], alpha=0.5)
#         plt.xlabel(column)
#         plt.ylabel(salary_column)
#         plt.title(f'Salary vs {column}')
#         plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(x[["job_role","department","Post"]])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),["job_role","department","Post"]),remainder="passthrough")
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
import pickle
```

```python
pickle.dump(pipe,open("SalaryPredictor.pkl","wb"))
```
